### Step 4: Display Data

#### Using pandas 

```python
all_data.plot()
```

#### Using Matplotlib (plt) 

```python
fig=plt.figure()
fig.set_size_inches(15, 10)
```

```python
ax1=fig.add_subplot(311)
ax2=fig.add_subplot(312)
ax3=fig.add_subplot(313)

```

```python
ax1.plot(all_data['celsius'], 'g-')
ax1.set_ylabel('C')
ax2.plot(all_data['ppm'], 'r-')
ax2.set_ylabel('PPM')
ax3.plot(all_data['eduroam'], 'm.')
ax3.set_ylabel('#users')
               
```

```python
fig
```

#### Using plotly.express

```python
# Using plotly.express
#import plotly.express as px
from plotly import express as px
fig = px.line(all_data, x=all_data.Date, y=[all_data.celsius,all_data.eduroam, all_data.ppm])
fig.show()
```

#### Using seaborn

```python
import seaborn as sns

sns.lineplot(data = all_data)
```
